```python
# Exploração dos Dados
Neste notebook vamos:
- Carregar o dataset
- Ver distribuição das classes
- Medir comprimentos das mensagens
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("SMSSpamCollection", sep="\t", names=["label","text"])
df.head()
```

```python
class_counts = df['label'].value_counts()
print(class_counts)
```

```python
class_counts.plot(kind='bar')
plt.title('Distribuição de ham vs spam')
plt.xlabel('Classe')
plt.ylabel('Número de mensagens')
plt.show()
```

```python
df['length'] = df['text'].str.len()
print(df['length'].describe())
```

```python
df['length'].hist(bins=30)
plt.title('Distribuição do comprimento das mensagens')
plt.xlabel('Números de Caracteres')
plt.ylabel('Frequência')
plt.show()
```
